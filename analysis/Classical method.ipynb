```python
import numpy as np
import pandas as pd
from fitting_data import gauss_est
from matplotlib import pyplot as plt
```

```python
data = np.load("data_2020-02-18_11-02.npy", allow_pickle = True)
```

```python

plt.plot(np.mod(np.array(data[1][0])/(1180-30)/2/3.1415, 1), data[0][0][0], ".")
plt.plot(np.mod(np.array(data[1][0])/(1180-30)/2/3.1415, 1), data[0][0][1], ".")
plt.plot(np.mod(np.array(data[1][0])/(1180-30)/2/3.1415, 1), data[0][0][2], ".")
plt.show()
```

```python
amplitudes = np.array(data[0][0])[:3, :]
phases = np.mod(np.array(data[1][0])/(1180-30)/2/3.1415, 1)
```

```python
amplitudes.shape
```

```python
model = gauss_est(0.005)
```

```python
model.__doc__
```

```python
train_ampl = amplitudes[:, :int(amplitudes.shape[1]*0.7)]
test_ampl = amplitudes[:, int(amplitudes.shape[1]*0.7):]
train_phases = phases[:int(len(phases)*0.7)]
test_phases = phases[int(len(phases)*0.7):]
```

```python
model.fit([train_phases, train_phases, train_phases], train_ampl)
```

```python
predict = model.predict(test_ampl)
```

```python
plt.figure(figsize = (20, 10))
plt.plot(test_phases, predict[0], ".")
plt.plot(test_phases, test_phases)
```
